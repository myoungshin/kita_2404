```python
# ì—¬ê¸° ëŒë¦¬ì„¸ìš”
import pymysql
import time

# ë°‘ì— ì—ëŸ¬ ëœ¨ëŠ”ê±´ ë¦¬ë·°ê°€ ë„ˆë¬´ ê¸¸ê±°ë‚˜ ë¹ˆì¹¸ì´ì–´ì„œ ê·¸ë ‡ìŠµë‹ˆë‹¤ ë¬´ì‹œí•˜ì…”ë„ ë©ë‹ˆë‹¤ Error for row 20363: ['ë´‰í”¼ì–‘ ìš©ì‚°ì  ê¹Œì§€ ë‚˜ì˜´
def insert_data_with_skip(data_to_insert):
    # MySQL ë°ì´í„°ë² ì´ìŠ¤ ì—°ê²° ì„¤ì •
    connection = pymysql.connect(host='localhost',
                                 user='kita3',
                                 passwd='kita3',
                                 db='kita3_db',
                                 charset='utf8mb4',  # Ensure utf8mb4 charset is used
                                 use_unicode=True)

    cursor = connection.cursor()

    # SQL ì¿¼ë¦¬ ì •ì˜
    insert_query = """
    INSERT INTO web (ShopName, ShopRating, Category_catchtable, Location_catchtable, 
    LunchDinner, Description, Reviews, Address, Nickname, User_Rating, MealTime, Date, 
    User_Review, Save, Category_catchtable_1, avg_ì ì‹¬, avg_ì €ë…, min_ì ì‹¬, max_ì ì‹¬, 
    min_ì €ë…, max_ì €ë…, Location, Lunch, Dinner) 
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """

    for idx, row in enumerate(data_to_insert):
        try:
            cursor.execute(insert_query, row)
            connection.commit()
        except pymysql.OperationalError as e:
            if e.args[0] == 1205:  # Lock wait timeout
                print(f"Lock wait timeout exceeded for row {idx}: {row}")
                connection.rollback()  # rollback the transaction for this row
            else:
                print(f"Operational error for row {idx}: {row} - {e}")
                connection.rollback()  # rollback the transaction for this row
        except Exception as e:
            print(f"Error for row {idx}: {row} - {e}")
            connection.rollback()  # rollback the transaction for this row


# Use the function to insert data with skipping problematic rows
insert_data_with_skip(data_to_insert)
```
