```python
import pandas as pd
import pymysql

# CSV 파일 열기
file_path = 'data_0716.csv'
df = pd.read_csv(file_path)

# NaN 값을 다른 값으로 대체
df[['min_점심', 'max_점심', 'min_저녁', 'max_저녁']] = df[['min_점심', 'max_점심', 'min_저녁', 'max_저녁']].fillna(0)
df = df.fillna('')  # NaN을 빈 문자열로 대체할 예시
# 'save' 컬럼의 값에서 쉼표 제거
df['Save'] = df['Save'].str.replace(',', '')

# 정수형으로 변환
df['Save'] = df['Save'].astype(int)

# 필요한 컬럼만 선택
data_to_insert = df[['ShopName', 'ShopRating', 'Category_catchtable', 'Location_catchtable',
       'LunchDinner', 'Description', 'Reviews', 'Address', 'Nickname',
       'User Rating', 'Meal Time', 'Date', 'User Review', 'Save',
       'Category_catchtable_1', 'avg_점심', 'avg_저녁', 'min_점심', 'max_점심',
       'min_저녁', 'max_저녁', 'Location', 'Lunch', 'Dinner']].values.tolist()

```

```python
# 여기 돌리세요
import pymysql
import time

# 밑에 에러 뜨는건 리뷰가 너무 길거나 빈칸이어서 그렇습니다 무시하셔도 됩니다 Error for row 20363: ['봉피양 용산점 까지 나옴
def insert_data_with_skip(data_to_insert):
    # MySQL 데이터베이스 연결 설정
    connection = pymysql.connect(host='localhost',
                                 user='kita3',
                                 passwd='kita3',
                                 db='kita3_db',
                                 charset='utf8mb4',  # Ensure utf8mb4 charset is used
                                 use_unicode=True)

    cursor = connection.cursor()

    # SQL 쿼리 정의
    insert_query = """
    INSERT INTO web (ShopName, ShopRating, Category_catchtable, Location_catchtable, 
    LunchDinner, Description, Reviews, Address, Nickname, User_Rating, MealTime, Date, 
    User_Review, Save, Category_catchtable_1, avg_점심, avg_저녁, min_점심, max_점심, 
    min_저녁, max_저녁, Location, Lunch, Dinner) 
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """

    for idx, row in enumerate(data_to_insert):
        try:
            cursor.execute(insert_query, row)
            connection.commit()
        except pymysql.OperationalError as e:
            if e.args[0] == 1205:  # Lock wait timeout
                print(f"Lock wait timeout exceeded for row {idx}: {row}")
                connection.rollback()  # rollback the transaction for this row
            else:
                print(f"Operational error for row {idx}: {row} - {e}")
                connection.rollback()  # rollback the transaction for this row
        except Exception as e:
            print(f"Error for row {idx}: {row} - {e}")
            connection.rollback()  # rollback the transaction for this row


# Use the function to insert data with skipping problematic rows
insert_data_with_skip(data_to_insert)
```
