Task3_0529. 주어진 DataFrame에서 'Category'별로 그룹화하여 각 그룹의 'Value' 열의 합계, 평균, 최대값, 최소값을 계산하세요.

Task7_0529. 주어진 DataFrame의 특정 열에 대해 map을 사용하여 등급을 부여하는 함수를 적용하여 변환을 수행하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Score": [85, 90, 95, 80, 75]
})
```

```python
# 주어진 dataframe:
print("원본 :\n",df)

# 특정 열에 map  사용하여 등급 부여하는 함수 적용, 변환을 수행
# score 옆에 열 추가, apply 사용
# 등급: 90 A, 80 B, 70 C, 60 이하는 D

df1=df.copy() # 원본 카피해서 사용

df1['grade']=df1['Score'].apply(lambda x:'A' if x>=90 else 'B' if x>=80 else 'C' if x>=70 else 'D')
print("\n등급 추가 : \n",df1)
```

Task14_0529. 주어진 데이터프레임에서 각 그룹별로 Value 열을 표준화하여 새로운 열로 추가하여 출력하세요.(apply, transform 두가지 적용)

```python
import pandas as pd

# 예제 데이터프레임 생성
data = {
    'Category': ['A', 'A', 'B', 'B', 'A', 'B', 'B', 'A', 'B', 'A'],
    'Value': [10, 15, 10, 20, 10, 30, 50, 10, 20, 15]
}

df = pd.DataFrame(data)
```

```python
print("Original DataFrame:\n",df)

# df에서 각 그룹별로 value열을 표준화하여 (.std()) 새로운 열로 추가, 출력하기
# 기존 형태 유지 필요. 그렇다면 사용자 함수로 표준화 필요
# 표준화:
def z_score(x):
    return (x-x.mean())/x.std()

# transform 사용:
df['표준화']=df.groupby('Category')['Value'].transform(z_score)
print("\ntransform 사용 : \n",df)

# apply 사용:
df['표준화']=df.groupby('Category')['Value'].apply(lambda x:(x-x.mean())/x.std()).reset_index(drop=True)
# df['표준화']=df.groupby('Category')['Value'].apply(z_score) # <== apply 뒤에는 사용자 함수를 못쓰는건지? 오류 뜸
print("\napply 사용 : \n",df)

```
