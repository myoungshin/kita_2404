Task1_0528. 연산메소드 이용 및 NaN 값을 0으로 대체 후 사칙연산을 수행한 후 결과를 출력하세요.

```python
import pandas as pd
import numpy as np

a=pd.Series({'국어':88,'영어':45,'수학':100,'과학':79, '미술':91})
b=pd.Series({'국어':70,'영어':100,'수학':89,'과학':92, '음악':100})

add1=a.add(b,fill_value=0)
sub1=a.sub(b,fill_value=0)
mul1=a.mul(b,fill_value=0)
div1=a.div(b,fill_value=0)

df=pd.DataFrame([add1,sub1,mul1,div1],index=['더하기','빼기','곱하기','나누기'])

print(df)
```

Task2_0528. 주어진 DataFrame의 N열에서 Series s값을 빼고 결과를 새로운 열 O에 저장 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "M": [15, 25, 35, 45, 55],
    "N": [100, 200, 300, 400, 500]
})
print(df,'\n')
# Series 생성
s = pd.Series([5, 10, 15, 20, 25])
```

```python
print("Original DataFrame : \n",df)

# df의 N열에서 Series s값을 빼기
O=df['N']- s
print("\ndf['N']-s : \n",O)

# 위 계산의 결과를 새로운 열 O에 저장, 출력
df['O']=O
print("\nNew DataFrame with column O added : \n",df)
```

Task3_0528. 주어진 DataFrame의 여러 열에 대해 각기 다른 Series를 더하고, 결과를 새로운 DataFrame으로 반환한 후 각 행의 합계를 계산하여 새로운 열에 추가하여 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [10, 20, 30, 40, 50],
    "C": [100, 200, 300, 400, 500]
})

# Series 생성
s1 = pd.Series([5, 5, 5, 5, 5])
s2 = pd.Series([10, 10, 10, 10, 10])
s3 = pd.Series([15, 15, 15, 15, 15])
```

```python
print("Original : \n",df)
# 기존 데이터프레임의 각 열에 각기 다른 Series 더하기
df_1=pd.Series(df['A']+s1)
df_2=pd.Series(df['B']+s2)
df_3=pd.Series(df['C']+s3)

# 더해서 구한 Series로 새 데이터프레임 만들기
new_df=pd.concat([df_1,df_2,df_3],axis=1)
print("\n새로 만든 데이터프레임 : \n", new_df)


# 새 데이터프레임에각 행의 합계를 계산해서 새로운 열 추가
new_df['합계']=new_df[0]+new_df[1]+new_df[2]
print("\n 합계 포함한 새 데이터프레임 : \n", new_df)
```

Task4_0528. df에서 세 열의 값을 더하여 새로운 컬럼을 생성한 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [10, 20, 30, 40, 50],
    "B": [1, 2, 3, 4, 5],
    "C": [5, 10, 15, 20, 25],
    "D": [3, 6, 9, 12, 15],
    "E": [2, 4, 6, 8, 10]
})
```

```python
print("원본 : \n",df)
# 세 열의 값을 더하여 새로운 열 생성, 출력
df1=df.copy()

df1['A+B+E']=df1['A']+df1['B']+df1['E']
print("\n세 열의 값을 더한 행 추가 : \n",df1)
```

Task5_0528. df에서 세 열의 값을 평균하여 새로운 컬럼을 생성 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [100, 200, 300, 400, 500],
    "B": [10, 20, 30, 40, 50],
    "C": [1, 2, 3, 4, 5],
    "D": [7, 8, 9, 10, 11],
    "E": [3, 6, 9, 12, 15]
})
```

```python
print("원본 : \n",df)

df['mean(A,B,C)']=df[['A','B','C']].mean(axis=1)
print("\n평균 추가 : \n",df)
```
