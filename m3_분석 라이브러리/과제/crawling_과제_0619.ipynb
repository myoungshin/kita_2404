Task1_0619. 다음 사항을 수행하세요.
- 첫 번째로 매칭되는 'p' 태그 찾기
- 모든 'p' 태그 찾기
- 클래스가 'content'인 첫 번째 'p' 태그 찾기
- 클래스가 'content'인 모든 'p' 태그 찾기
- 특정 'p' 태그의 모든 부모 태그 찾기
- 특정 'p' 태그의 첫 번째 부모 태그 찾기
- 특정 'p' 태그의 다음 형제 태그 찾기
- 특정 'p' 태그의 이전 형제 태그 찾기
- 특정 'p' 태그 다음에 위치한 모든 태그나 문자열 찾기
- 특정 'p' 태그 이전에 위치한 모든 태그나 문자열 찾기 

```python
from bs4 import BeautifulSoup

html_content='<html><body><h1>Title</h1><p class="content">First paragraph.</p><p class="content">Second paragraph.</p></body></html>'
soup=BeautifulSoup(html_content, 'html.parser')
print(soup.prettify())
```

```python
# - 첫 번째로 매칭되는 'p' 태그 찾기
first_p=soup.find('p')
print("첫 번째로 매칭되는 'p' 태그 찾기 : \n", first_p)

# - 모든 'p' 태그 찾기
all_p=soup.find_all('p')
print("\n모든 'p' 태그 찾기 : ", all_p)

# - 클래스가 'content'인 첫 번째 'p' 태그 찾기
content_first_p=soup.select_one('.content')
print("\n클래스가 'content'인 첫 번째 'p' 태그 찾기 : \n", content_first_p)

# - 클래스가 'content'인 모든 'p' 태그 찾기
content_all_p=soup.select('.content')
print("\n클래스가 'content'인 모든 'p' 태그 찾기 : \n", content_all_p)

# - 특정 'p' 태그의 모든 부모 태그 찾기
first_p_all_parents=first_p.find_parents()
print("\n특정 'p' 태그의 모든 부모 태그 찾기 : \n", first_p_all_parents)

# - 특정 'p' 태그의 첫 번째 부모 태그 찾기
first_p_first_parent=first_p.find_parent()
print("\n특정 'p' 태그의 첫 번째 부모 태그 찾기 : \n", first_p_first_parent)

# - 특정 'p' 태그의 다음 형제 태그 찾기
first_p_next_sib=first_p.find_next_sibling()
print("\n특정 'p' 태그의 다음 형제 태그 찾기 : \n", first_p_next_sib)

# - 특정 'p' 태그의 이전 형제 태그 찾기
first_p_prev_sib=first_p.find_previous_sibling()
print("\n특정 'p' 태그의 이전 형제 태그 찾기 : \n", first_p_prev_sib)

# - 특정 'p' 태그 다음에 위치한 모든 태그나 문자열 찾기
first_p_next_element=first_p.find_all_next()
print("\n특정 'p' 태그 다음에 위치한 모든 태그나 문자열 찾기 : ")
for all in first_p_next_element:
    print(all)

# - 특정 'p' 태그 이전에 위치한 모든 태그나 문자열 찾기 
first_p_prev_element=first_p.find_all_previous()
print("\n특정 'p' 태그 이전에 위치한 모든 태그나 문자열 찾기  : ")
for all in first_p_prev_element:
    print(all)
```

Task2_0619. ID를 이용해서 '네이버 뉴스' 추출하세요.

```python
import requests
from bs4 import BeautifulSoup

# 예시 네이버 뉴스 페이지 URL
url='https://news.naver.com'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response=requests.get(url, headers=headers)
response
```

```python
soup=BeautifulSoup(response.content, 'html.parser')
print(soup.prettify())
```

```python
naver_news=soup.find(id="browserTitleArea")
print(naver_news.text)
```
