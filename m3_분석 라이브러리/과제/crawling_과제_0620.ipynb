Task1_0620. 네이버 영화 순위 사이트에서 영화제목, 가격, 타입(구매 or 대여) 정보를 가져와서 TITLE, PRICE, TYPE 3개의 컬럼과 100개의 데이터포인트로 구성된 데이터프레임을 출력하세요.

```python
from bs4 import BeautifulSoup
import requests as rq
import pandas as pd
import re

headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"}
url='https://serieson.naver.com/v3/movie/ranking/realtime'
response=rq.get(url, headers=headers)
print(response)
text=response.text

soup=BeautifulSoup(text, 'html.parser')

# << 영화제목, 가격, 타입 (구매 or  대여) 정보 가져오기 >>

# 1. title:
# <span class="Title_title__s9o0D">인사이드 아웃(패키지상품 : 더빙판 + 부가영상 추가증정)</span>
title_list=soup.find_all('span', class_='Title_title__s9o0D')
titles=[title.get_text() for title in title_list]
print(titles)

# 2. price:
# <span class="Price_price__GqXqo">7,150<svg class="PriceCashIcon_root__8U3by Price_cash_icon__qbTPC" xmlns="http://www.w3.org/2000/svg" viewBox="0 0 12 12" fill="#777" aria-hidden="true"><path class="st0" d="M6,12c-3.3,0-6-2.7-6-6s2.7-6,6-6s6,2.7,6,6S9.3,12,6,12z M6,1C3.2,1,1,3.2,1,6s2.2,5,5,5s5-2.2,5-5S8.8,1,6,1z M6.1,9C7.8,9,8.9,8.1,9,6.8V6.8H7.4l0,0c-0.1,0.6-0.6,1-1.3,1C5.2,7.8,4.7,7.1,4.7,6v0c0-1.1,0.5-1.8,1.4-1.8c0.7,0,1.2,0.4,1.3,1 l0,0H9V5.2C8.9,3.9,7.7,3,6.1,3C4.2,3,3,4.1,3,6v0C3,7.9,4.2,9,6.1,9z"></path></svg><span class="blind">캐시</span></span>
price_list=soup.find_all('span', class_='Price_price__GqXqo')
prices=[price.get_text() for price in price_list]
prices_rev=[price.replace('캐시', '') for price in prices]
print(prices_rev)

# 3. type (구매/대여):
# <span class="Price_text__pRk_f">구매</span>
type_list=soup.find_all('span', class_='Price_text__pRk_f')
types=[type.get_text() for type in type_list]
print(types)


# # << title, price, type 으로 구성된 데이터프레임 만들기 >>
df_titles=pd.DataFrame({
    'title': titles,
    'price': prices_rev,
    'type': types,
})
df_titles

```

Task3_0620. 앞에서 출력한 기사 리스트를 컬럼은 인덱스 번호와 텍스트 2개 컬럼으로 구성하여 pandas 데이터프레임으로 변환 후 csv 파일로 저장 후 다시 불러오세요.

```python
# 블러오기
import requests as rq
import re
import pandas as pd
from bs4 import BeautifulSoup

headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"}
url='https://news.daum.net/politics#1'
r=rq.get(url, headers=headers)
html=r.text
soup=BeautifulSoup(html, 'html.parser')

lines=soup.select('body > div > main > section > div > div > ul > li > div > div > strong > a')

# 기사 리스트
articles=[article.get_text() for article in lines]
print(articles)

# 인덱스 리스트
index_num=[]
for i in range(len(articles)):
    index_num.append(i)
print(index_num)

# 데이터프레임 변환 (index, text 컬럼명으로 사용)
df_articles=pd.DataFrame({
    'index':index_num,
    'text': articles})

df_articles['index']=df_articles['index']+1
df_articles=df_articles.set_index('index')
print(df_articles)

# csv 파일로 저장 & 다시 불러오기 
df_articles.to_csv('articles.csv')
```
