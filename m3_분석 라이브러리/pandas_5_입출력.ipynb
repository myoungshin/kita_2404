```python
# csv 파일 읽고 저장하기
import numpy as np
import pandas as pd
ar1 = np.arange(10).reshape(2,5) # 2행 5열짜리 0-9까지의 데이터 생성
df = pd.DataFrame(ar1,columns=list('abcde'))
df.to_csv('test.csv') # 데이터프레임을 csv 로 저장
df.to_csv('test.csv',index=None) # index=None 으로 저장 시 파일 불러올 때 별도로 index_col= 지정 안해도 이상하게 안나옴
```

```python
df=pd.read_csv('test.csv',index_col=0)
df
```

```python
df1=df.copy()
```

```python
df1.to_csv('test1.csv')
df1=pd.read_csv('test1.csv',index_col=0)
df1
```

```python
# json 파일 읽고 쓰기
df1.iloc[0]=2
df1.iloc[1]=3
df1
```

```python
# json 파일 읽고 쓰기
df1.to_json('test2.json')
df_j=pd.read_json('test2.json')
df_j
```

```python
# 바이너리 파일 읽고 쓰기
df1.to_pickle('test3.pkl')
df1=pd.read_pickle('test3.pkl')
df1
```
