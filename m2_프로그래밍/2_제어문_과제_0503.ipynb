##### add_order() 함수 부분이 풀리지 않아서 GPT의 솔루션 확인함

```python
# GPT가 수정해준 코드 :
# 1. 주문판 출력
def display_menu():
    print(
'''
KITA 식당에 오신 것을 환영합니다

========================================

          - 메  뉴 -
    1. 불고기 비빔밥 : 12,000원
    2. 야채 비빔밥 : 8,000원
    3. 전주 비빔밥 : 10,000원

    세트 주문시 : 3000원 추가
    (세트는 밥과 반찬이 추가됩니다.)

========================================
'''
)

# 2. 주문 접수를 받는 함수
def order():
    menu = {1: ['불고기 비빔밥', 12000], 2: ['야채 비빔밥', 8000], 3: ['전주 비빔밥', 10000]}
    set_price = 3000
    global current_order
    global current_price
    menu_select = int(input("원하시는 메뉴의 번호를 선택하세요. 1,2,3 >>> "))
    current_order = menu[menu_select]
    print(f"선택하신 메뉴는 {current_order[0]}입니다.")

    # 세트 여부 확인
    menu_set = input("세트로 하시겠습니까? y,n >>> ").lower()
    if menu_set == 'y':
        current_price = current_order[1] + set_price
        print(f"세트를 선택하셨습니다. 금액은 {current_price}원 입니다.")
    else:
        current_price = current_order[1]
        print(f"단품을 선택하셨습니다. 금액은 {current_price}원 입니다.")

# 3. 추가 주문 여부를 확인
def add_order(final_order, final_price):
    while True:
        menu_add = input("메뉴를 추가하시겠습니까? y,n >>> ").lower()
        if menu_add == 'y':
            order()  # 주문 추가
            final_order.append({'메뉴': current_order[0], '가격': current_price})  # 주문 내역에 추가
            final_price += current_price  # 총 가격 업데이트
            print(f"현재까지 선택하신 메뉴는 : {final_order}입니다.")
        else:
            print(f"최종적으로 선택하신 메뉴는 : {final_order}입니다.")
            print(f"최종 가격은 {final_price}원 입니다.")
            break

# 이 모든 것을 한 번에 해결해줄 최종보스:
def main():
    final_order = []  # 주문 내역을 저장할 리스트
    final_price = 0  # 총 가격 초기화

    while True:
        display_menu()
        order()
        final_order.append({'메뉴': current_order[0], '가격': current_price})  # 초기 주문 내역 저장
        final_price += current_price  # 초기 가격 계산
        add_order(final_order, final_price)  # 추가 주문 확인
        reset_order = input("다시 주문하시겠습니까? y,n >>> ")
        if reset_order.lower() == 'n':
            break
    print("주문해주셔서 감사합니다. 잠시만 기다려주세요.")

if __name__ == "__main__":
    main()
```

```python
# GPT 수정 사항 확인 후 업데이트한 최종본
# 여러개 음식을 동시에 주문할 수 있는 주문 시스템을 작성하세요. (구조 지향)

# 1. 주문판을 보여주는 함수
# 2. 주문 접수를 받는 함수
# 3. 추가 주문 여부를 확인
# 4. 추가 주문 선택 시 2~3 다시 실행
# 5. 추가 주문 없을 시 가격 안내


# 1. 주문판 출력
def display_menu():
    print(
'''
KITA 식당에 오신 것을 환영합니다

========================================

          - 메  뉴 -
    1. 불고기 비빔밥 : 12,000원
    2. 야채 비빔밥 : 8,000원
    3. 전주 비빔밥 : 10,000원

    세트 주문시 : 3000원 추가
    (세트는 밥과 반찬이 추가됩니다.)

========================================
'''
)

# 2. 주문 접수를 받는 함수
def order():
# a. 메뉴 선택하세요
    menu={1:['불고기 비빔밥', 12000], 2:['야채 비빔밥', 8000], 3:['전주 비빔밥', 10000]}
    set_price=3000
    global current_order
    menu_select=int(input("원하시는 메뉴의 번호를 선택하세요. 1,2,3 >>> "))
    current_order=menu[menu_select]
    print(f"선택하신 메뉴는 {current_order[0]}입니다.")

# b. 세트로 하시겠습니까?
    menu_set=input("세트로 하시겠습니까? y,n >>> ").lower()
    global current_price
    if menu_set=='y':
        current_price=current_order[1]+set_price
        print(f"세트를 선택하셨습니다. 금액은 {current_price}원 입니다.")
    else:
        current_price=current_order[1]
        print(f"단품을 선택하셨습니다. 금액은 {current_price}원 입니다.")

# 3. 추가 주문 여부를 확인
def add_order(final_order, final_price):
    while True:
        menu_add=input("메뉴를 추가하시겠습니까? y,n >>> ").lower()
        if menu_add=='y':
            order()
            final_order.append({'메뉴': current_order[0], '가격':current_price}) # order()에서 입력된 다중의 current order의 cumulative 리스트를 원하는데 어떻게 해야하는지 모르겠네
            final_price+=current_price
            print(f"현재까지 선택하신 메뉴는 : {final_order}입니다.")
        else:
            print(f"최종적으로 선택하신 메뉴는 : {final_order}입니다.")
            print(f"최종 가격은 {final_price}원 입니다.")
            break

# 이 모든 것을 한 번에 해결해줄 최종보스:

def main():
    final_order=[]
    final_price=0
    while True:
        display_menu()
        order()
        final_order.append({'메뉴': current_order[0], '가격': current_price})  # 초기 주문 내역 저장
        final_price += current_price  # 초기 가격 계산
        add_order(final_order, final_price)
        reset_order=input("다시 주문하시겠습니까? y,n >>> ")
        if reset_order.lower()=='n':
            break
    print("주문해주셔서 감사합니다. 잠시만 기다려주세요.")

if __name__ == "__main__":
    main()






```
