Task1_0502. 짝수와 짝수를 입력하면 곱한 값을 출력하고 홀수와 홀수를 입력하면 덧셈 값을 출력하고 그외는 다시 입력하라는 메시지를 출력하세요.

```python
# 사용자에게 수 2개를 받기
# case 1: 둘 다 짝수 = num1*num2
# case 2: 둘 다 홀수 = num1+num2
# case 3: 그 외 케이스는 다시 입력

while True:
    num1=int(input("첫번째 숫자를 입력하세요 : "))
    num2=int(input("두번째 숫자를 입력하세요 : "))
    if num1%2==0 and num2%2==0:
        print(f'{num1},{num2}를 입력하셨습니다. 입력하신 두 수의 곱은 {num1*num2}입니다')
        break
    elif num1%2!=0 and num2%2!=0:
        print(f'{num1},{num2}를 입력하셨습니다. 입력하신 두 수의 합은 {num1+num2}입니다')
        break
    else:
        print("다시 입력하세요")
```

Task2_0502. 현재 계절 구분 프로그램

시나리오 : 3 ~ 5월은 봄, 6 ~ 8월은 여름, 9 ~ 11월은 가을, 12 ~ 2월은 겨울로 구분. 지금 계절을 알려주는 프로그램
각 계절별 출력 멘트는 계절에 맞게 창작
if 조건문으로 datetime 모듈을 import해서 datetime.now 함수를 사용해서 작성

```python
spring=[3,4,5]
summer=[6,7,8]
autumn=[9,10,11]
winter=[12,1,2]

import datetime
import pytz
tz=pytz.timezone('Asia/Seoul')
now=datetime.datetime.now(tz)
now_month=now.month
now_month # 현재 달

if now_month in spring:
    print(f"{now_month}월입니다. 현재는 봄입니다")
elif now_month in summer:
    print(f"{now_month}월입니다. 현재는 여름입니다")
elif now_month in autumn:
    print(f"{now_month}월입니다. 현재는 가을입니다")
else:
    print(f"{now_month}월입니다. 현재는 겨울입니다")
```

Task3_0502. 1부터 99까지 아래와 같이 2개의 수를 곱해서 가장 큰 수를 구하세요.

1*99, 2*98 ...99*1

```python
# 1~99까지
# 인덱스 기준 양끝에서 중간으로, 교차해서 다시 양끝으로 가는 형태로 숫자 하나씩 설렉트
# 각 뽑힌 수는 한쌍
# 각 쌍의 곱을 비교
# 각 쌍을 곱했을 때 그 곱이 가장 큰 값 출력

max_mult=0
mult=0
for i in range(1,100):
    mult=i*(100-i)
    if max_mult<mult:
        max_mult=mult
print(max_mult)
```
