
Task3_0426. 주어진 숫자 리스트에서 특정 값보다 큰 요소가 하나라도 존재하는지 검사하고, 그 결과를 불 값으로 반환하는 함수를 작성하세요.

numbers = [1, 2, 3, 10, 20]
threshold = 15

```python
numbers = [1, 2, 3, 10, 20]

def compare_num(threshold):
    for number in numbers:
        if threshold<number:
            return bool(number)
    return False

threshold = 15 # <== 특정값
print(compare_num(threshold))
```

```python
# 강사님 풀이 1 - if문으로 사용

if any(num>15 for num in numbers):
    print("True")
else:
    print("False")


# 강사님 풀이 2 - 함수 사용
def check_greater_than_threshold(numbers,threshold):
    for num in numbers:
        if num>threshold:
            return True
    return False

numbers = [1, 2, 3, 10, 20]
threshold = 15
result=check_greater_than_threshold(numbers,threshold)
print(result)
```

Task4_0426. 주어진 문자열이 특정 문자열을 포함하는지 확인하고 결과를 불 값으로 반환하는 함수를 작성하세요.

text = "hello world"
substring = "world"

```python
text = "hello world"
substring = "world" # <== 특정 문자열

print(substring in text)
```

```python
# 강사님 풀이
def contains_substring(text, substring):
    if substring in text:
        return True
    else:
        return False

text = "hello world"
substring = "world"
result=contains_substring(text,substring)
print(result)
```

Task5_0426. 주어진 연도가 윤년인지 판별하는 함수를 작성하세요. 윤년은 다음의 조건을 만족해야 합니다:

4로 나누어 떨어진다.
100으로 나누어 떨어지지 않거나, 400으로 나누어 떨어지면 윤년이다.
year = 2020

True

```python
year = 2020

def leap_year(year):
    #조건1
    if year%4==0 and year%100!=0 or year%400==0:
        return True
    else:
        return False
print(leap_year(2020))
```

```python
# 강사님 풀이

def is_leap_year(year):
    if year%4==0 and (year%100!=0 or year%400==0):
        return True
    else:
        return False
year=2020
result=is_leap_year(year)
print(result)
```

```python
class Calculator:
    result=0
    def addition(self, value1, value2):
        result=value1+value2
        return result

    def subtraction(self, value1, value2):
        result=value1-value2
        return result

    def multiplication(self, value1, value2):
        result=value1*value2
        return result

    def division(self, value1, value2):
        result=value1/value2
        return result

nums=Calculator()
print(f"덧셈 : {nums.addition(10,20)}")
print(f"뺄셈 : {nums.subtraction(10,20)}")
print(f"곱셈 : {nums.multiplication(10,20)}")
print(f"나눗셈: {nums.division(10,20)}")
```

```python
#  강사님 풀이
class Calculator:
    result=0
    def add(self, n1, n2):
        return n1+n2

    def sub(self, n1, n2):
        return n1-n2

    def mul(self, n1, n2):
        return n1*n2

    def div(self, n1, n2):
        if n2==0:
            return "Error : Division by zero"
        else:
            return n1/n2

obj1=Calculator()
print(obj1.add(1,1))
print(obj1.sub(1,1))
print(obj1.mul(1,1))
print(obj1.div(1,1))
```

Task7_0426. 다음 과제를 수행하세요.

사용자로 부터 텍스트를 입력 받는다.
문자을 단어 단위로 분리 : split()
단어의 빈도수를 저장할 딕셔너리를 생성
각 단어의 빈도 수를 계산(for 문 / if else문): word=word.lower()
결과 출력
[ 예시 ]

문장을 입력하세요: I love apple. I love orange. Apple is tasty

{'i': 2, 'love': 2, 'apple.': 1, 'orange.': 1, 'apple': 1, 'is': 1, 'tasty': 1}

```python
# 사용자로 부터 텍스트를 입력 받는다.
# msg1='I love apple. I love orange. Apple is tasty'
msg1=input("문장을 입력하세요 : ")

# 단어를 모두 소문자로 받기 위해 word=word.lower() 사용
msg1_lower=msg1.lower()

# 문자를 단어 단위로 분리 : split()
msg1_lower.split()
msg1_txt=msg1_lower.split()

# 단어의 빈도수를 저장할 딕셔너리를 생성
msg1_count={}

# # # # 각 단어의 빈도 수를 계산(for 문 / if else문): word=word.lower()
for i in msg1_txt:
    if i in msg1_count:
        msg1_count[i]+=1
    else:
        msg1_count[i]=1

# # 결과 출력
print(msg1_count)

```

```python
# 강사님 풀이

text=input("문장을 입력하세요 : ")
words=text.split()
word_frequency={}
for word in words:
    word=word.lower() # 모두 소문자화 시키기 위해 .lower() 사용.
    if word in word_frequency:
        word_frequency[word]+=1
    else:
        word_frequency[word]=1

print(word_frequency)
```
