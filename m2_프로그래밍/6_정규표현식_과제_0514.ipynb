Task1_0514. BC, CC, ABC 모두 C가 출력되는 정규 표현식을 ( )에 작성하세요.

```python
import re

pattern=re.compile('C') # <== compile() 안을 작업해야함.
text1='BC'
text2='CC'
text3='ABC'
p1=pattern.search(text1)
p2=pattern.search(text2)
p3=pattern.search(text3)
# print(p1)
# print(p2)
# print(p3)
print(f"'BC'에서의 C 출력: {p1.group()}, {p1}")
print(f"'CC'에서의 C 출력: {p2.group()}, {p2}")
print(f"'ABC'에서의 C 출력: {p3.group()}, {p3}")
```

```python
# 강사님 풀이:
import re
pattern=re.compile('A?C')
text1='BC'
text2='CC'
text3='ABC'
p1=pattern.search(text1)
p2=pattern.search(text2)
p3=pattern.search(text3)
# print(p1)
# print(p2)
# print(p3)
print(f"'BC'에서의 C 출력: {p1.group()}, {p1}")
print(f"'CC'에서의 C 출력: {p2.group()}, {p2}")
print(f"'ABC'에서의 C 출력: {p3.group()}, {p3}")
```

```python
# 스핀오프

import re

pattern=re.compile('A?C')
test_strings=['AC', 'C','ABC','AAC','BCC']
for s in test_strings:
    match=pattern.match(s)
    if match:
        print(f'"{s}" found a match!')
    else:
        print(f'"{s}" no match')
```

Task2_0514. ' Python3 is very good programming language!'에서  ['Python', 'is', 'very', 'good', 'programming', 'language']를 모두 출력하세요.

```python
import re

text1=' Python3 is very good programming language!'
pattern=re.compile('\w+')
p1=pattern.findall(text1)
print(p1)
```

```python
# 강사님 풀이
text1=' Python3 is very good programming language!'
pattern=re.compile('[a-zA-Z]+')
p1=pattern.findall(text1)
print(p1)
```
