#### Q. 주어진 문자열 리스트에서 유효한 Python 변수 이름만을 추출하여 반환하는 함수를 작성하세요

```python
identifiers=["var1","2things","variable_name", "time!"]

# Python에서 사용 불가능한 변수명:
# 1. keyword
# 2. int/float으로 시작
# 3. 공백 있음
# 4. _ 외 특수문자 사용

# 즉, Python 에서 사용 가능한 변수명은 아래의 조건을 갖춰야 하며, 함수 또한 하기 조건에 부합하는 경우 identifiers 에서 요소 추출되도록 만들어야 함
# 1. 첫 글짜가 숫자가 아니라면 통과
# 2. 특수 문자 중에서도 _ 는 사용 가능하니 _ 있는 경우도 통과



def valid_python_variables(identifiers):
    valid_identifiers = []
    for identifier in identifiers:
        # 첫 글자가 숫자인지 확인
        if identifier[0].isdigit():
            continue
        # 숫자, 영문자, 밑줄로만 구성되어 있는지 확인
        if all(char.isalnum() or char == '_' for char in identifier):
            valid_identifiers.append(identifier)
    return valid_identifiers

identifiers = ["var1", "2things", "variable_name", "time!"]
valid_variables = valid_python_variables(identifiers)
print(valid_variables)
```
