Task1_0425. 주어진 숫자 리스트에서 최소값과 최대값을 찾아 출력하세요.

numbers = [58, 45, 69, 19, 4, 87, 29, 13, 39, 15]

```python
numbers = [58, 45, 69, 19, 4, 87, 29, 13, 39, 15]

print(max(numbers)) # 최대
print(min(numbers)) # 최소
```

```python
# 강사님 풀이 1

# numbers = [58, 45, 69, 19, 4, 87, 29, 13, 39, 15]

# min_value=min(numbers)
# max_value=max(numbers)
# print(f"최소값: {min_value}, 최대값 : {max_value}")


# 강사님 풀이 2 (내장함수 사용 X)

numbers = [58, 45, 69, 19, 4, 87, 29, 13, 39, 15]

# 최소, 최대값을 첫 번째 요소로 초기화
min_value=numbers[0] # numbers의 첫 요소인 58을 넣고
max_value=numbers[0]

for number in numbers: # 리스트의 모든 요소를 순회
    if number<min_value: # 현재의 요소가 현재 최소값보다 작으면 최소값을 갱신 (첫 요소로 들어간 58을 기준으로 이후 요소들을 차례대로 비교하면서 더 작은 값을 min_value로 계속 대체해 감)
        min_value=number # 위 조건에 맞다면 좌측 내용 실행, 아니라면 넘어가기
    if number>max_value: # 현재의 요소가 현재 최대값보다 크면 최대값을 갱신
        max_value=number
print(f"최소값: {min_value}, 최대값: {max_value}")




```

Task2_0425. 주어진 숫자 리스트의 모든 요소의 합계와 평균을 계산하고 출력하세요

numbers = [58, 45, 69, 19, 4, 87, 29, 13, 39, 15]

```python
numbers = [58, 45, 69, 19, 4, 87, 29, 13, 39, 15]

# numbers 합계 :
total=0 # 합계 변수 생성해주고
for i in numbers: # numbers에 있는 각 요소들이
    total+=i # sum 변수에 더해지도록

# numbers 평균값 :
average=total/len(numbers) # 합계를 numbers 요소 개수만큼 나눠주기

print(f"합계 : {total}, 평균: {average}")
```

```python
# 강사님 풀이1

numbers = [58, 45, 69, 19, 4, 87, 29, 13, 39, 15]
# 합계 계산
total=sum(numbers)
#평균 계산
average=total/len(numbers)
#결과 출력
print(f"합계: {total}, 평균: {average}")


# 강사님 풀이2 - numpy 사용

import numpy as np # numpy 라이브러리를 불러오는데, 'np'라는 별칭으로 부르겠음
numbers = [58, 45, 69, 19, 4, 87, 29, 13, 39, 15]
print(f"합계: {sum(numbers)}, 평균: {np.mean(numbers)}")


# 강사님 풀이 3 - statistics 사용하여 평균값 구하기

import statistics
numbers = [58, 45, 69, 19, 4, 87, 29, 13, 39, 15]
average=statistics.mean(numbers)
print(f"평균: {np.mean(numbers)}")
```

Task4_0425. 주어진 리스트에서 연속해서 반복되는 요소만 제거하고, 결과 리스트를 반환하세요. 단, 처음 등장하는 요소는 유지해야 합니다.

```python
items = ['apple', 'banana', 'cherry', 'apple', 'cherry', 'apple']

# list -> set -> list 자료형 변환으로:
items_set=set(items)
# items_rev=list(items_set) # TypeError <== 이거 왜 오류나는지?
items_rev=[i for i in items_set]
print(items_rev) # 근데 set를 쓰니까 첫번째 apple이 사라짐. set에서 중복값 삭제는 뒤에서부터 아닌지...?


# for문 사용:

items_rev2=[]
for item in items:
    if item not in items_rev2:
        items_rev2.append(item)
print(items_rev2)
```

```python
# 강사님 풀이 1
li=['a', 'a', 'b', 'c', 'c', 'c', 'd', 'e', 'e']
result=[li[0]] # result 리스트를 만들어주는데, 담을 요소는 li리스트의 첫번 째 요소 ('a')
for i in li[1:]: # li 리스트의 2번째 요소부터 끝까지를 보는데
    if i !=result[-1]: # 각 요소가 result 리스트의 가장 끝 요소와 같지 않다면
        result.append(i) # result 리스트에 해당 요소 추가
print(result)


# 강사님 풀이 2 - 사용자 함수 생성하여 풀기
def remove_consecutive_duplicates(items): # 본 함수에 매개변수 items를 넣으면, 아래 기능을 실행
    if not items: # items가 없다면
        return [] # 빈 리스트를 반환하기 (items에 값이 없어서 하기 for문 내용만 실행 시 에러 발생. if not items 넣으면 에러 발생시 용이.)
    result=[items[0]] # 본 함수 실행 후 새로 생성한 result 리스트에 넣어주는데, 일단 result 리스트에는 items의 첫 요소를 담아주기
    for current in items[1:]: # items  리스트의 2번째 요소부터 끝까지 순회하고, 각 회차마다 맞이하는 현 요소가 (current):
        if current !=result[-1]: # result  리스트의 가장 끝 요소와 같지 않다면
            result.append(current) # result 리스트에 현 요소를 추가해주기
        return result # 순회가 끝나면 result 반환
print(result)
```
