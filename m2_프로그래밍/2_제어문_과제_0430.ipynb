```python
num1=int(input("팩토리얼 계산할 숫자를 입력하세요 : "))
# 5x4x3x2x1 표현을 어떻게 하지?
```

```python
# 강사님 풀이
num1=int(input("팩토리얼 계산할 숫자를 입력하세요 : "))
factorial=1
for i in range(1,num1+1):
    factorial*=i
print(f"{num1}의 팩터리얼은 {factorial}입니다")
```

```python
# GPT 풀이 (팩토리얼 계산 표현까지 포함)
num1 = int(input("팩토리얼 계산할 숫자를 입력하세요: "))
factorial = 1
calculation = ''

for i in range(1, num1 + 1):
    factorial *= i
    calculation += str(i)
    if i != num1:
        calculation += ' x '

print(f"{num1}! = {calculation} = {factorial}")
```

Task7_0430. 주어진 리스트에서 최대값을 찾아 출력하세요. numbers=[34,78,2,45,99,23]

```python
numbers=[34,78,2,45,99,23]

# 최간단히 한다면:
print(max(numbers))

# for문 사용한다면:
max_number=0
for number in numbers:
    if max_number<number:
        max_number=number
print(max_number)
```

Task8_0430. 1부터 10 사이의 임의의 숫자를 맞추는 게임을 만드세요. 사용자가 숫자를 맞출 때까지 입력을 계속 받으며, 정답을 맞추면 게임을 종료하세요. 단, 입력 숫자가 정답보다 큰지 작은지 힌트를 주는 방식으로 코드 작성

```python
# 1부터 10 사이의 임의의 숫자를 맞추는 게임
import random
x=random.randint(1,10)
# 사용자가 숫자를 맞출 때까지 입력을 계속 받으며, 정답을 맞추면 게임을 종료하세요

while True:
    number=int(input("숫자를 입력하세요 : "))
    if x==number:
        print("숫자를 맞추셨습니다")
        break
    else:
        print("틀렸습니다. 다시 맞춰보세요.\n힌트: ") # "입력 숫자가 정답보다 큰지 작은지 힌트를 주는 방식으로 코드 작성" <== 이거 어떻게 하지?
```

```python
# 강사님 풀이
import random
target_number=random.randint(1,10)
while True:
    guess=int(input("1부터 10 사이의 숫자를 맞춰보세요 : "))
    if guess>target_number:
        print("더 작은 숫자입니다!")
    elif guess < target_number:
        print("더 큰 숫자 입니다!")
    else:
        print("정답입니다!")
        break
```
