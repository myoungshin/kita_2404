Task7_0430. 주어진 리스트에서 최대값을 찾아 출력하세요. numbers=[34,78,2,45,99,23]

```python
numbers=[34,78,2,45,99,23]

# 최간단히 한다면:
print(max(numbers))

# for문 사용한다면:
max_number=0
for number in numbers:
    if max_number<number:
        max_number=number
print(max_number)
```
